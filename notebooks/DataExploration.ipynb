```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
# from gensim.models import KeyedVectors

from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score
from sklearn.metrics import f1_score, roc_auc_score, confusion_matrix, roc_curve
```

```python
field = 'transcription'
tfidf_vectorizer = TfidfVectorizer()
tfidf_vectorizer.fit(train_data[field].values)
# print(list(tfidf_vectorizer.get_feature_names_out()))
print(len(list(tfidf_vectorizer.get_feature_names_out())))
train_feature_set=tfidf_vectorizer.transform(train_data[field].values)
test_feature_set=tfidf_vectorizer.transform(val_data[field].values)
```

```python
train_feature_set
```

```python
test_feature_set
```

```python
rf_model = RandomForestClassifier(n_estimators=200,
                                  criterion='entropy',
                                  max_depth=4,
                                  class_weight='balanced',
                                  random_state=0,
                                  verbose=0)

rf_model.fit(train_feature_set, train_data['lab_act'])
```

```python
def get_predictions(model, X_test, is_prob=False):
  if is_prob:
    return model.predict_proba(X_test)[:,1]
  # get predicted labels
  preds = model.predict(X_test)
  return preds


def print_metrices(Y_test, Y_pred):
  # GET ACCURACY, PRECISION, RECALL, F1-SCORE
  print("ACCURACY: {}".format(round(accuracy_score(Y_test, Y_pred), 4)))

  print("ROC AUC SCORE: {}".format(round(roc_auc_score(Y_test, Y_pred), 4)))

  print("BULLISH PRECISION: {}".format(round(precision_score(Y_test, Y_pred), 4)))

  print("BULLISH RECALL: {}".format(round(recall_score(Y_test, Y_pred), 4)))

  print("BULLISH F1-SCORE: {}".format(round(f1_score(Y_test, Y_pred), 4)))

  inverse_y_test = np.where(Y_test==1, 0, 1)
  inverse_y_pred = np.where(Y_pred==1, 0, 1)
  print("BEARISH PRECISION: {}".format(round(precision_score(inverse_y_test, 
                                                             inverse_y_pred), 4)))

  print("BEARISH RECALL: {}".format(round(recall_score(inverse_y_test, inverse_y_pred), 4)))

  print("BEARISH F1-SCORE: {}".format(round(f1_score(inverse_y_test, inverse_y_pred), 4)))

  total_bullish = sum(confusion_matrix(Y_test, Y_pred)[1])
  total_bearish = sum(confusion_matrix(Y_test, Y_pred)[0])

  true_pred_bullish = confusion_matrix(Y_test, Y_pred)[1][1]
  true_pred_bearish = confusion_matrix(Y_test, Y_pred)[0][0]
  print("{} bullish samples are correctly predicted out of {}".format(true_pred_bullish,
                                                                      total_bullish))
  print("{} bearish samples are correctly predicted out of {}".format(true_pred_bearish,
                                                                      total_bearish))
```
